```python
from PIL import Image
import requests
from io import BytesIO
from IPython.display import SVG,display
```

```python
url_imagem = 'https://dinaproductions.com/wp-content/uploads/2017/02/Aya-BE.jpg'

# Lendo a imagem
response = requests.get(url_imagem)
img = Image.open(BytesIO(response.content))

# Exibindo imagem
display(img)
                      
```

```python
# Varrer a matriz de pixels e clarear a imagem
for i in range(img.size[0]):
    for j in range(img.size[1]):
        pixel = img.getpixel((i,j))
        img.putpixel((i,j),(pixel[0]*2,pixel[1]*2,pixel[2]*2))

# Exibindo imagem
display(img)
```

```python
# Varrendo pixels
for x in range(img.width):
    for y in range(img.height):
        pixel = img.getpixel((x,y))
        # Alterando o pixel
        pixel = (pixel[0] + 100, pixel[1] + 100, pixel[2] + 100)
        img.putpixel((x,y), pixel)

# Exibindo imagem
display(img)
```
